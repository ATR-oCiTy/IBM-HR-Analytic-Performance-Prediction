```python
# In the database, there exists some categorical values, which needs to be converted to numerical values.

from sklearn.preprocessing import LabelEncoder
```

```python
encode = LabelEncoder()
```

```python
df['BusinessTravel'] = encode.fit_transform(df['BusinessTravel'].astype('str'))
```

```python
df['Department'] = encode.fit_transform(df['Department'].astype('str'))
```

```python
df['EducationField']=encode.fit_transform(df['EducationField'].astype('str'))
```

```python
df['Gender'] = encode.fit_transform(df['Gender'].astype('str'))
```

```python
df['MaritalStatus'] = encode.fit_transform(df['MaritalStatus'].astype('str'))
```

```python
df['Over18'] = encode.fit_transform(df['Over18'].astype('str'))
```

```python
df['OverTime'] = encode.fit_transform(df['OverTime'].astype('str'))
```

```python
def converter(attrition):
    if attrition == 'Yes':
        return 1
    else :
        return 0
```

```python
df['Attrition1']=df['Attrition'].apply(converter)
```

```python
df.head()
```

```python
df.drop('Attrition',axis=1,inplace=True)
```

```python
# Initializing X and y variables for train-test-split.
# X variable consists of the Independent Variables and y variable consists of the Dependent Variable.

X=df.drop('Attrition1',axis=1)
y=df['Attrition1']
```

```python
X.head()
```

```python
y.head()
```

```python
# Dividing the data to train data and test data

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
# Using Support Vector Machine algorithm

from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
# Using HyperParameter Tuning to tune the parameters in SVM.

from sklearn.model_selection import GridSearchCV
```

```python
param_grid={'C':[1,0.1,10,12,13,23,56,100,145,200,300,1000,0.01],'gamma':[1,0,0.1,0.5,0.75,0.01,0.001,0.0001,10]}
```

```python
grid = GridSearchCV(SVC(),param_grid,verbose=3,refit=True)
```

```python
# Fitting X_train , y_train to the grid object.

grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
# Predicting X_test using this

pred = grid.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,pred))
print('\n')
print(classification_report(y_test,pred))
```

```python
# We can see that this model has an accuracy score of 84 but has 0 precision in predicting 1(Yes). 
# This is due to the less number of data for the 'Yes' attrition.
# One way of fixing this is using SMOTE(Synthetic Minority Oversampling Technique).
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE()
```

```python
# Creating a copy of the dataset

X_sm,y_sm=smote.fit_sample(X,y)
```

```python
X_train_sm,X_test_sm,y_train_sm,y_test_sm=train_test_split(X_sm,y_sm,test_size=0.3,random_state=101)
```

```python
grid.fit(X_train_sm,y_train_sm)
```

```python
grid.best_params_
```

```python
pred = grid.predict(X_test_sm)
```

```python
print(confusion_matrix(y_test_sm,pred))
print('\n')
print(classification_report(y_test_sm,pred))
```

```python
# This model has an accuracy of 74, but prediction of 0 is low.
```
